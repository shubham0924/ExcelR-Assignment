```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
movies = pd.read_csv("C:\\Users\\INDIA\\Downloads\\Assignment csv file\\my_movies.csv")
movies
```

```python
movie = movies.iloc[:,5:]
movie
```

# 1.Association Rule with 10% support and 60% confidence :-

```python
dataset1 = apriori(movie, min_support=0.1, use_colnames=True)
dataset1.shape
```

```python
rule1 = association_rules (dataset1, metric='confidence', min_threshold=0.6)
rule1
```

```python
rule1 = rule1[rule1.lift>1]
rule1
```

# The number of rules we are getting are 134.

```python
rule1[['support','confidence','lift']].hist()
```

```python
plt.scatter(rule1.support, rule1.confidence)
plt.xlabel('Support')
plt.ylabel('Confidence')
```
