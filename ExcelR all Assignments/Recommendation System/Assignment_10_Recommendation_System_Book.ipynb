```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data = pd.read_csv("C:\\Users\\INDIA\\Downloads\\Assignment csv file\\book.csv")
data
```

```python
data.index = np.arange(1, len(data)+1)
```

```python
data['userId']= data.index
data
```

```python
books = data.iloc[:,:-1]
books
```

```python
# for calculating the distances between the user's
# we will use cosine similarity
```

```python
user_sim = 1- (pairwise_distances(books, metric='cosine'))
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index = data.userId
user_sim_df.columns = data.userId
```

```python
user_sim_df
```

```python
# diag0nal values are distance of same users 
# hence we will convert all the diagonal values into 0
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim_df
```

```python
# most similar users :-

user_sim_df.idxmax(axis=1)
```
